```python
# https://stackoverflow.com/questions/13314626/local-solar-time-function-from-utc-and-longitude
# https://www.esrl.noaa.gov/gmd/grad/solcalc/solareqns.PDF
def get_solar_time_and_angles(dt,longit,latit): 
    gamma = 2 * pi / 365 * (dt.timetuple().tm_yday - 1 + float(dt.hour - 12) / 24) # unit = radians
    eqtime = 229.18 * (0.000075 + 0.001868 * cos(gamma) - 0.032077 * sin(gamma) \
             - 0.014615 * cos(2 * gamma) - 0.040849 * sin(2 * gamma)) # unit = minutes
    decl = 0.006918 - 0.399912 * cos(gamma) + 0.070257 * sin(gamma) \
           - 0.006758 * cos(2 * gamma) + 0.000907 * sin(2 * gamma) \
           - 0.002697 * cos(3 * gamma) + 0.00148 * sin(3 * gamma) # unit = radians
    time_offset = eqtime + 4 * longit # unit = minutes
    tst = dt.hour * 60 + dt.minute + dt.second / 60 + time_offset # unit = minutes
    solar_time = datetime.combine(dt.date(), time(0)) + timedelta(minutes=tst) # unit = datetime
    solar_hour_angle = (tst/4) - 180 # unit = degrees
    solar_zenith_angle = 180/pi * acos(sin(latit*pi/180)*sin(decl) + \
                                       cos(latit*pi/180)*cos(decl)*cos(solar_hour_angle*pi/180)) # unit = degrees
    solar_alt_angle = 90 - solar_zenith_angle # unit = degrees
    #return solar_time, solar_hour_angle, solar_zenith_angle, solar_alt_angle
    return solar_time, solar_alt_angle
```

### Apply simple alg

```python
# - Find DVM depth w/ largest diff in Sv btwn noon and midnight AND
# DVM depth w/ largest Sv during noon + no pg<50% + no non-zero pflag
# + no depths examined below max_depth defined above  
noon_hrs = [12] # noon hours 
mn_hrs = [0] # midnight hours
all_days = pd.to_datetime(Sv['time'].values).normalize().unique()
tot_cdays = len(all_days)
sd_depth0 = [None]*tot_cdays
sd_depth1 = [None]*tot_cdays
daynow = pd.to_datetime(Sv['time'][0].values)
for iday in range(0,tot_cdays):
    daynow = all_days[iday]
    day_strnow = daynow.strftime('%Y-%m-%d') 
    Svnow = Sv.where((Sv_pg>pg_thresh) & (Sv_pflag==0)).sel(time=day_strnow)
    Svnow = Svnow.where(Sv_depth<=max_depth, drop=True)
    Sv_depth_arrnow = Sv_depth_arr[Sv_depth_arr<=max_depth]
    Sv_hrnow = Svnow.groupby('time.hour').mean(dim='time') 
    # - Compute dvm depth as largest Sv during noon
    if len(set(Sv_hrnow['hour'].values).intersection(noon_hrs))>0:
        Sv_noonnow = Sv_hrnow.sel(hour=noon_hrs).mean(dim='hour')
        sd_depth0[iday] = Sv_depth_arrnow[Sv_noonnow.argmax()]
    else:
        sd_depth0[iday] = np.nan
    # - Compute dvm depth as largest Sv diff btwn noon minus midnight
    if (len(set(Sv_hrnow['hour'].values).intersection(mn_hrs))>0
       ) & (len(set(Sv_hrnow['hour'].values).intersection(noon_hrs))>0):
        Sv_mnnow = Sv_hrnow.sel(hour=mn_hrs).mean(dim='hour')
        Sv_diffnow = Sv_noonnow-Sv_mnnow
        sd_depth1[iday] = Sv_depth_arrnow[Sv_diffnow.argmax()]
    else:
        sd_depth1[iday] = np.nan
    fig,axes = plt.subplots(nrows=1, ncols=3, figsize=(11,3))
    pc = axes[0].pcolormesh(Sv_hrnow['hour'].values, Sv_depth_arrnow, Sv_hrnow.transpose(), vmax=0)
    axes[0].set_ylim([max_depth,0]); axes[0].set_ylabel('Depth [m]')
    fig.colorbar(pc, ax=axes[0], orientation='vertical', pad=0.01)
    axes[0].set_title('day ' + str(iday) + ', ' + day_strnow)
    axes[1].plot(Sv_diffnow, Sv_depth_arrnow); axes[1].set_ylim([max_depth,0])
    axes[1].set_title('Noon-MN: ' + str(sd_depth1[iday]) + ' m')
    if len(set(Sv_hrnow['hour'].values).intersection(noon_hrs))>0:
        axes[2].plot(Sv_noonnow, Sv_depth_arrnow, label='Noon'); axes[2].set_ylim([max_depth,0])
    if len(set(Sv_hrnow['hour'].values).intersection(mn_hrs))>0:
        axes[2].plot(Sv_mnnow, Sv_depth_arrnow, label='MN'); axes[2].set_ylim([max_depth,0])
    axes[2].set_title('Noon max: ' + str(sd_depth0[iday]) + ' m')
    axes[2].legend()
    #print(sd_depth0[iday])
print(datetime.now().strftime("%H:%M:%S"))
```
